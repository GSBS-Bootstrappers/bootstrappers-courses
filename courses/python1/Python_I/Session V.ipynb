```python
import pysam

def getregions(filename, chr):
  count = 0
  f = pysam.Samfile(filename)
  chridx = f.gettid(chr)
  for i in f.fetch(chr):
    print(i.qname, i.is_reverse, i.alen, i.cigarstring)
    count = count + 1
    if count > 100:
        break

# main
if __name__ == "__main__":
  filename = "data/ENCFF001BQW.filtered.bam"
  chromosome = "chrX"
  getregions(filename, chromosome)
```
